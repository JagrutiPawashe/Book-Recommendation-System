# Popularity Based Recommender System

Here we marge the book and rating dataset for to make similar dataset

```python
rating_with_name=rating.merge(book,on='ISBN')
```

```python
rating_with_name.head()
```

There We check the number of the vote per book

```python
num_rating_df=rating_with_name.groupby('Book-Title').count()["Book-Rating"].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num of rating'},inplace=True)
num_rating_df
```

```python
avg_rating_df = rating_with_name.groupby('Book-Title')['Book-Rating'].agg(lambda x: x.astype(float).mean()).reset_index()
avg_rating_df.rename(columns = {'Book-Rating' : 'avg_rating'}, inplace = True)
avg_rating_df
```

```python
popularity_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popularity_df
```

```python
popularity_df[popularity_df['num of rating']>=50].sort_values('avg_rating',ascending=False).head(10)
```

```python
popularity_df=popularity_df.merge(book,on="Book-Title").drop_duplicates("Book-Title")[['Book-Title','Book-Author','Year-Of-Publication','Image-URL-S','avg_rating','num of rating',"avg_rating"]]
```

```python
popularity_df
```

# Collaborative Filtering Based Recommender System

```python
x=rating_with_name.groupby('User-ID').count()['Book-Rating']>200
vaild_user=x[x].index
```

```python
vaild_user
```

```python
filter_rating=rating_with_name[rating_with_name["User-ID"].isin(vaild_user)]
```

```python
filter_rating
```

```python
y=filter_rating.groupby("Book-Title").count()['Book-Rating']>=50
famous_book=y[y].index
```

```python
final_rating=filter_rating[filter_rating["Book-Title"].isin(famous_book)]
final_rating.drop_duplicates()
```

```python
final_rating_pivot=final_rating.pivot_table(columns='User-ID',index='Book-Title',values='Book-Rating')
final_rating_pivot
```

```python
final_rating_pivot.fillna(0,inplace=True)
final_rating_pivot
```

Calculate Distance for simlarity


```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(final_rating_pivot)
```

```python
similarity_score
```

```python
def recommend(book_name):
    index=np.where(final_rating_pivot.index==book_name)[0][0]
    similar_item=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:4]
    
    for i in similar_item:
        print(final_rating_pivot.index[i[0]])
    
  
```

```python
recommend("Zoya")
```
